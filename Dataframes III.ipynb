## A Review of the .apply() Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#define a costume python function
def convert_to_string_and_add_million(number):
    return str(number) + "MILLIONS"
```

```python
#apply the costum function to the column indicated
bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_million)
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_million)
```

```python
#or we can apply it to all the columns we want all together
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_million)
```

```python
bond.head(3)
```
