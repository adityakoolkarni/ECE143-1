```python
import pandas as pd
```

## The .unstack() Method, Part1

```python
world = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
#stack up the columns
s = world.stack()
s.head(3)
```

```python
#to unstack the stacked multidimension dataframe; it will unravel the most inner layer to the most right column
s.unstack()
```

```python
s.unstack().unstack()
```

```python
s.unstack().unstack().unstack()
```

## The .unstack() Method, Part2

```python
world = pd.read_csv("worldstats.csv", index_col=["country", "year"])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
#other than the default method to use .unstack()
# You can unstack based on the location of the layer on the multiindex stack
s.unstack(2)
```

```python
#we can also provide negative indeces; -1 and 2 columns would mean the same column here
s.unstack(-1)
```

## The .unstack() Method, Part3

```python
world = pd.read_csv("worldstats.csv", index_col=["country", "year"])
s = world.stack()
#here the most outer layer, in this country has an index layer of 0, population has an index level of 1, and so on
world.head(3)
```

```python
s.unstack(level=[1, 0]) #notice the difference with the next cell
```

```python
s.unstack(level = [0, 1])
```

```python
#can also use strings for them
s.unstack(level=["year", "country"])
```

```python
s.unstack(level=["country", "year"])
```

```python
#if there are any NAN and we want to clear it, we can use the fill_value parameter
s.unstack("year", fill_value = 0)
```
